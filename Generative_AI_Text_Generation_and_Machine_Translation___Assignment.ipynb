Question 3: What is the difference between Autoencoders and Variational
Autoencoders (VAEs) in the context of text generation?
sol) While both Autoencoders (AEs) and Variational Autoencoders (VAEs) use an encoder-decoder structure to compress and reconstruct data, they serve very different purposes in the context of generation.1. Standard Autoencoders (AEs)A standard Autoencoder is primarily a compression tool. It takes an input (like a sentence), compresses it into a fixed "bottleneck" vector (the latent space), and then tries to reconstruct the exact same sentence.The Latent Space: In a standard AE, the latent space is discrete and unconstrained. Each input is mapped to a single, specific point.The "Gap" Problem: Because the points are scattered randomly, the spaces between them are "empty." If you pick a random point between two known sentences in the latent space, the decoder will likely produce gibberish because it hasn't learned what that "gap" represents.Role in Text: AEs are great for dimensionality reduction or feature extraction, but they are poor at generating new text.2. Variational Autoencoders (VAEs)A VAE turns the latent space into a continuous probability distribution. Instead of mapping an input to a single point, it maps it to a mean ($\mu$) and a variance ($\sigma$).The Latent Space: The VAE forces the latent space to follow a specific distribution (usually a Gaussian/Normal distribution). This ensures the space is "packed" and continuous.Interpolation: Because the space is continuous, you can pick any random point in the latent space, and the decoder will be able to generate a plausible, coherent sentence. You can even "walk" from one sentence to another, seeing how the text smoothly evolves.Role in Text: VAEs are true generative models. They allow for controlled creative generation and variations of existing themes.Key Differences ComparisonFeatureAutoencoder (AE)Variational Autoencoder (VAE)Mapping TypeDeterministic (Input $\rightarrow$ Point)Probabilistic (Input $\rightarrow$ Distribution)Latent SpaceDiscontinuous (has "holes")Continuous and RegularizedPrimary UseCompression, DenoisingData Generation, Creative SynthesisMath ToolReconstruction LossReconstruction Loss + KL DivergenceWhy it matters for Text GenerationIn text, small changes in the latent space can lead to massive changes in meaning. A standard AE is too "brittle"—shifting a coordinate slightly might turn a coherent sentence into a string of random characters. A VAE provides the mathematical structure needed to navigate the "meaning" of language smoothly, making it possible to generate diverse and grammatically correct sentences that weren't in the original training set.

Question 4: Describe the working of attention mechanisms in Neural Machine
Translation (NMT). Why are they critical?
sol) In Neural Machine Translation (NMT), the Attention Mechanism is the engine that allows a model to "focus" on specific parts of a source sentence while generating each word of the target translation.Before attention, NMT models used a simple Encoder-Decoder structure where the entire source sentence was compressed into a single, fixed-length vector. This created a "bottleneck" where the model would often forget the beginning of long sentences by the time it reached the end.How Attention Works: Step-by-StepInstead of relying on one "thought vector," the attention mechanism allows the decoder to look back at all the hidden states of the encoder at every step of the translation.Alignment Scores: For every word the model is currently translating, it calculates a "score" for every word in the source sentence. This score represents how relevant a source word is to the current target word.Softmax Weights: These scores are passed through a Softmax function to turn them into probabilities (weights) that sum to 1.Example: If translating "The blue cat" to French ("Le chat bleu"), when the model is ready to output "bleu," the attention weight for "blue" will be very high (e.g., 0.9).Context Vector: The model creates a weighted sum of all the encoder's hidden states based on these weights. This "Context Vector" is a filtered version of the source sentence, highlighting only the most important information for the current word.Decoding: The decoder uses this specific context vector, along with its own previous state, to predict the next word in the translation.Why Attention is CriticalThe introduction of attention was a "paradigm shift" in AI for several reasons:Solving the Bottleneck: It removed the need to cram a 50-word sentence into a single vector. The model now has a "searchable memory" of the source text.Handling Long-Range Dependencies: Traditional RNNs (Recurrent Neural Networks) struggle with long sentences due to the "vanishing gradient" problem. Attention provides a direct mathematical link between distant words.Handling Word Order (Alignment): Languages have different structures (e.g., Adjective-Noun vs. Noun-Adjective). Attention allows the model to "jump around" the source sentence to find the right word, regardless of its position.Explainability: Attention maps provide a visual way to see what the AI was "thinking." By looking at the weights, researchers can see exactly which source words triggered a specific translated word.Comparison: Standard NMT vs. Attention-Based NMTFeatureStandard Encoder-DecoderAttention-Based NMTInformation FlowFixed-length vector (The Bottleneck)Dynamic Context VectorSentence LengthPerformance drops significantly on long textMaintains high accuracy on long sentencesComplexity$O(n)$ (Sequential processing)$O(n^2)$ (Pairs every word with every other word)AlignmentImplicit and often poorExplicit and highly accurate

Question 6: Use the following small text dataset to train a simple Variational
Autoencoder (VAE) for text reconstruction:
["The sky is blue", "The sun is bright", "The grass is green",
"The night is dark", "The stars are shining"]
1. Preprocess the data (tokenize and pad the sequences).
2. Build a basic VAE model for text reconstruction.
3. Train the model and show how it reconstructs or generates similar sentences.
Include your code, explanation, and sample outputs
sol) To build a Variational Autoencoder (VAE) for text, we have to bridge the gap between discrete words and a continuous latent space. Since we are using a very small dataset, we will use a Character-level or Word-level embedding approach.

For this example, I'll use a word-level tokenizer and a basic GRU-based VAE.

1. Preprocessing the Data
We need to map words to integers and ensure all sequences are the same length.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, losses, models
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Dataset
data = ["The sky is blue", "The sun is bright", "The grass is green",
        "The night is dark", "The stars are shining"]

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(data)
sequences = tokenizer.texts_to_sequences(data)
vocab_size = len(tokenizer.word_index) + 1
max_len = max(len(s) for s in sequences)

# Padding
x_train = pad_sequences(sequences, maxlen=max_len, padding='post')
```

2. Building the VAE Model
The VAE consists of an Encoder that outputs a distribution (mean and variance), a Sampling layer, and a Decoder.

```python
latent_dim = 2  # Small latent space for visualization

# --- ENCODER ---
encoder_inputs = layers.Input(shape=(max_len,))
x = layers.Embedding(vocab_size, 8)(encoder_inputs)
x = layers.GRU(16)(x)
z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)

# Sampling Layer
def sampling(args):
    z_mean, z_log_var = args
    epsilon = tf.random.normal(shape=tf.shape(z_mean))
    return z_mean + tf.exp(0.5 * z_log_var) * epsilon

z = layers.Lambda(sampling)([z_mean, z_log_var])

# --- DECODER ---
decoder_inputs = layers.Input(shape=(latent_dim,))
x = layers.RepeatVector(max_len)(decoder_inputs)
x = layers.GRU(16, return_sequences=True)(x)
decoder_outputs = layers.TimeDistributed(layers.Dense(vocab_size, activation="softmax"))(x)

# Instantiate models
encoder = models.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
decoder = models.Model(decoder_inputs, decoder_outputs, name="decoder")

# --- VAE LOSS (Reconstruction + KL Divergence) ---
class VAE(models.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super(VAE, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder

    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)

            # Loss Calculation
            recon_loss = tf.reduce_mean(
                tf.reduce_sum(losses.sparse_categorical_crossentropy(data, reconstruction), axis=1)
            )
            kl_loss = -0.5 * tf.reduce_mean(
                tf.reduce_sum(1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var), axis=1)
            )
            total_loss = recon_loss + kl_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        return {"loss": total_loss, "recon_loss": recon_loss, "kl_loss": kl_loss}

vae = VAE(encoder, decoder)
vae.compile(optimizer='adam')
```

3. Training and Reconstruction
Because the dataset is tiny, we'll overfit slightly to see if it can reconstruct the original sentences.

```python
# Train for 500 epochs for convergence on tiny data
vae.fit(x_train, epochs=500, verbose=0)

# Reconstructing a sentence
def reconstruct(sentence):
    seq = tokenizer.texts_to_sequences([sentence])
    padded = pad_sequences(seq, maxlen=max_len, padding='post')
    _, _, z = encoder.predict(padded)
    preds = decoder.predict(z)

    # Convert indices back to words
    result = []
    for word_probs in preds[0]:
        idx = np.argmax(word_probs)
        if idx != 0:
            result.append(tokenizer.index_word[idx])
    return " ".join(result)

print(f"Original: {data[0]}")
print(f"Reconstructed: {reconstruct(data[0])}")
```

Sample Output & Explanation
Reconstruction: After training, if you input "The sky is blue", the VAE should output the same or a very similar sequence like "The sky is blue".

Generation: If you sample a random point in the latent space (e.g., np.random.normal(size=(1, 2))), the decoder might produce a "hybrid" sentence like "The sky is bright"—demonstrating that it has learned a continuous representation of your specific vocabulary.

Why this works: The Embedding layer learns the relationships between words, while the GRU handles the sequential nature of the sentences. The KL Divergence loss ensures that "sky" and "sun" end up near each other in the latent space because they appear in similar sentence structures.

Question 8: Implement a simple attention-based encoder-decoder model for
English-to-Spanish translation using Tensorflow or PyTorch.


```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# Configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
input_vocab_size = 5000  # English
output_vocab_size = 5000 # Spanish
hidden_dim = 256
embedding_dim = 128

class Encoder(nn.Module):
    def __init__(self, vocab_size, emb_dim, hid_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, emb_dim)
        self.gru = nn.GRU(emb_dim, hid_dim, batch_first=True)

    def forward(self, x):
        embedded = self.embedding(x)
        output, hidden = self.gru(embedded)
        return output, hidden

class BahdanauAttention(nn.Module):
    def __init__(self, hid_dim):
        super().__init__()
        self.Wa = nn.Linear(hid_dim, hid_dim)
        self.Ua = nn.Linear(hid_dim, hid_dim)
        self.Va = nn.Linear(hid_dim, 1)

    def forward(self, query, values):
        # query: (batch, 1, hid_dim), values: (batch, seq_len, hid_dim)
        score = self.Va(torch.tanh(self.Wa(query) + self.Ua(values)))
        weights = F.softmax(score, dim=1)
        context = torch.bmm(weights.transpose(1, 2), values)
        return context, weights

class DecoderWithAttention(nn.Module):
    def __init__(self, vocab_size, emb_dim, hid_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, emb_dim)
        self.attention = BahdanauAttention(hid_dim)
        self.gru = nn.GRU(hid_dim + emb_dim, hid_dim, batch_first=True)
        self.fc = nn.Linear(hid_dim, vocab_size)

    def forward(self, x, hidden, encoder_outputs):
        # x: (batch, 1) - current word
        embedded = self.embedding(x)

        # Calculate attention
        query = hidden.transpose(0, 1) # (batch, 1, hid_dim)
        context, weights = self.attention(query, encoder_outputs)

        # Combine context and embedded word
        rnn_input = torch.cat((embedded, context), dim=2)
        output, hidden = self.gru(rnn_input, hidden)

        prediction = self.fc(output.squeeze(1))
        return prediction, hidden, weights

# --- Dummy Inference Test ---
def test_inference():
    # Mock inputs (Batch size=1, Sequence length=5)
    english_input = torch.randint(0, input_vocab_size, (1, 5)).to(device)

    encoder = Encoder(input_vocab_size, embedding_dim, hidden_dim).to(device)
    decoder = DecoderWithAttention(output_vocab_size, embedding_dim, hidden_dim).to(device)

    # 1. Encode
    enc_out, enc_hid = encoder(english_input)

    # 2. Start Decoding with <SOS> token (assume index 1)
    dec_input = torch.tensor([[1]]).to(device)
    dec_hid = enc_hid

    # 3. Generate one word
    prediction, next_hid, attn_weights = decoder(dec_input, dec_hid, enc_out)

    print("--- Model Output Shapes ---")
    print(f"Prediction Probabilities: {prediction.shape}") # (Batch, Vocab)
    print(f"Attention Weights: {attn_weights.shape}")      # (Batch, SeqLen, 1)

test_inference()
```
